# Problem Statement :

__Customer satisfaction has emerged as one of the most important factors that guarantee the success of online store; it has been posited as a key stimulant of purchase, repurchase intentions and customer loyalty. A comprehensive review of the literature, theories and models have been carried out to propose the models for customer activation and customer retention. Five major factors that contributed to the success of an e-commerce store have been identified as: service quality, system quality, information quality, trust and net benefit. The research furthermore investigated the factors that influence the online customers repeat purchase intention. The combination of both utilitarian value and hedonistic values are needed to affect the repeat purchase intention (loyalty) positively. The data is collected from the Indian online shoppers. Results indicate the e-retail success factors, which are very much critical for customer satisfaction.__

# Numeric Columns :

# Univariate Analysis :
<font color=orange>DISTRIBUTION PLOT FOR NUMERIC COLUMNS</font>

```python
for i in df[numeric_columns]:
    plt.figure()
    sns.distplot(df[i], color="k")
```

_From the above distribution plot we can clearly visualize the high density categories for each feature_

<font color=orange>SWARMPLOT FOR NUMERIC COLUMNS</font>

```python
for i in df[numeric_columns]:
    plt.figure()
    sns.swarmplot(df[i], color='red')
```

_With swarmplot visualization technique, we are able to view the spread of values in this continuous dataset. Swarm-plot spreads out the data points of the variable automatically to avoid overlap._

_From here we can observe that most of the datapoints are evenly and closely spread, and hence less chances of outliers._

<font color=orange>BOXPLOT FOR NUMERIC COLUMNS</font>

```python
df[numeric_columns].plot(kind='box', subplots=True, layout=(8,5), figsize=(20,20), colormap='YlOrRd_r')
```

_Outliers can be observed in some columns like 'Pincode', 'Payment', 'Abandon_Frequency', 'Content_Feasibility', 'Similar_ProductComparison', 'Product_InfoClarity' etc_

<font color=orange>COUNTPLOT FOR OBJECT COLUMNS</font>

```python
for i in df[object_columns]:              
    plt.figure(figsize = (10, 5))
    sns.countplot(df[i])
    plt.xticks(rotation=90)
    plt.title(i)
    plt.show()
```
